```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df1['combined_features'] = df1['ClusterTitle'] + " " + df1['PathTitle'] + " " + df1['SOCTITLE']
df1['Normalized_Rating'] = (df1['Rating'] - df1['Rating'].min()) / (df1['Rating'].max() - df1['Rating'].min())
```

```python
vectorizer = TfidfVectorizer()
feature_matrix = vectorizer.fit_transform(df1['combined_features'])
similarity_matrix = cosine_similarity(feature_matrix)
```

```python
def recommend(index, top_n=5, rating_threshold=0.3):
    similarity_scores = list(enumerate(similarity_matrix[index]))
    similarity_scores = [(i, score) for i, score in similarity_scores if df1.loc[i, 'Normalized_Rating'] >= rating_threshold]
    combined_scores = [(i, score + 0.1 * df1.loc[i, 'Normalized_Rating']) for i, score in similarity_scores]
    combined_scores = sorted(combined_scores, key=lambda x: x[1], reverse=True)
    recommended_indices = [i[0] for i in combined_scores[1:top_n+1]]

    return df1.loc[recommended_indices, ['Name', 'Mail', 'ClusterTitle', 'PathTitle', 'SOCTITLE', 'Rating']]
```

```python
recommendations = recommend(index=0)
recommendations = recommendations.reset_index(drop=True)
```
